```python
pho = pd.read_csv("../input/total-data/pho.csv")
```

```python
pho.describe()
```

```python
pho = pho.dropna(axis = "rows", how = 'any')
title = pd.DataFrame(pho,columns=['Title'])
comment = pd.DataFrame(pho,columns=['Comment'])
camera = pd.DataFrame(pho,columns=['Camera'])
pho
```

```python
plt.figure(figsize=(10,12))

# Add title
plt.title("number comment of stores")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(y=pho.Title[:50], x=pho.Comment[0:50])

# Add label for vertical axis
plt.ylabel("display stores")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=comment[0:100],palette=['red'])

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=camera[:100])

# Add label for horizontal axis
plt.xlabel("Stores")
```

```python
sns.regplot(x=comment, y=camera)
```
